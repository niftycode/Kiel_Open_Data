```python
import io
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

URL = "https://www.kiel.de/opendata/kiel_kultur_freizeit_sport_tourismus_gaeste_uebernachtungen.csv"

# Set width and height
plt.rcParams["figure.figsize"] = (14.0, 8.0)

# Fetch data
csv_data = requests.get(URL).content

# Create dataframe
df = pd.read_csv(io.StringIO(csv_data.decode("utf-8")), sep=";")

# Create a subplot
fig, ax = plt.subplots()

# Drop NA values
df_cleaned = df.dropna(how="all")

# Unsigned integer (0 to 65535)
df_years = df_cleaned["Jahr "].astype(np.uint16)

x = df_years.values
y = df_cleaned["Uebernachtungen Sommerhalbjahr"].values

# Set title and x/y label
plt.title("Übernachtungen in Kiel\n (1988 - 2019)", size="x-large")
plt.ylabel("Übernachtungen Sommerhalbjahr", size="x-large")
plt.xlabel("Jahr", size="x-large")

# Plot data
plt.plot(y, "*-", markersize=6, linewidth=1, color="g")

ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation="vertical")

plt.show()




```
