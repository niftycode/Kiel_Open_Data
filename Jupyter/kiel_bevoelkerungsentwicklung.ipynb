```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20.0, 10.0)

# create subplots
fig, ax = plt.subplots()

# read csv file
df = pd.read_csv("data/kiel_bevoelkerung_flaeche_gebietsstand.csv", sep=";")

# clean data
df_cleaned = df.dropna(how="all")
df_years = df_cleaned['Jahr'].astype(np.uint16)

# assign two colums
x = df_years.values
y = df_cleaned["Einwohner insgesamt"].values

plt.title("Bev√∂lkerungsentwicklung in Kiel", size="x-large")
plt.ylabel("Anzahl", size="x-large")
plt.xlabel("Jahr", size="x-large")

plt.plot(y, "*-", markersize=6, linewidth=1, color="b")

ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation="vertical")

plt.show()
```
