# Garbage in Kiel

(Household Garbage and Industrial Waste, 1999-2019)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# create subplots
fig, ax = plt.subplots()

# read csv file
df = pd.read_csv("data/kiel_garbage.csv", sep=";")

# clean data
df_cleaned = df.dropna(how="all")
df_years = df_cleaned["Jahr"].astype(np.uint16)

# assign two colums
x = df_years.values
y = df_cleaned["Haus- und Geschaeftsabfall (in to.)"].values

plt.title("Garbage in Kiel", size="x-large")
plt.ylabel("Tons", size="x-large")
plt.xlabel("Year", size="x-large")

plt.plot(y, "o-", markersize=6, linewidth=1, color="b")

ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation="vertical")

plt.show()
```
